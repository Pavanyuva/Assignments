**Implementing Clustering Algorithms:**

•	Implement the K-Means, hierarchical, and DBSCAN algorithms using a programming language such as Python with libraries like scikit-learn or MATLAB.

•	Apply each clustering algorithm to the pre-processed dataset to identify clusters within the data.

•	Experiment with different parameter settings for hierarchical clustering (e.g., linkage criteria), K-means (Elbow curve for different K values) and DBSCAN (e.g., epsilon, minPts) and evaluate the clustering results


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Elbow Method to find optimal K
sse = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)

plt.plot(k_values, sse, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters K')
plt.ylabel('SSE')
plt.show()

# Fit K-Means with optimal K
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_features)

# Calculate silhouette score
silhouette_kmeans = silhouette_score(scaled_features, kmeans_labels)
print(f'Silhouette Score for K-Means: {silhouette_kmeans:.2f}')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Perform hierarchical clustering
linked = linkage(scaled_features, 'ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram')
plt.show()

# Fit hierarchical clustering (let’s assume we choose 3 clusters)
hierarchical = AgglomerativeClustering(n_clusters=optimal_k)
hierarchical_labels = hierarchical.fit_predict(scaled_features)

# Calculate silhouette score
silhouette_hierarchical = silhouette_score(scaled_features, hierarchical_labels)
print(f'Silhouette Score for Hierarchical Clustering: {silhouette_hierarchical:.2f}')
```

```python
from sklearn.cluster import DBSCAN

# Experiment with different epsilon and min_samples
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_features)

# Calculate silhouette score
if len(set(dbscan_labels)) > 1:

    core_samples_mask = dbscan_labels != -1
    silhouette_dbscan = silhouette_score(scaled_features[core_samples_mask], dbscan_labels[core_samples_mask])
    print(f'Silhouette Score for DBSCAN: {silhouette_dbscan:.2f}')
else:
    print('DBSCAN did not form any clusters.')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_features)
plt.figure(figsize=(8, 6))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

**Cluster Analysis and Interpretation:**

•	Analyse the clusters generated by each clustering algorithm and interpret the characteristics of each cluster. Write you insights in few comments.


```python
# Analyze clusters for K-Means
df['kmeans_cluster'] = kmeans_labels
kmeans_cluster_analysis = df.groupby('kmeans_cluster').mean()
print("K-Means Cluster Analysis:\n", kmeans_cluster_analysis)

# Analyze clusters for Hierarchical
df['hierarchical_cluster'] = hierarchical_labels
hierarchical_cluster_analysis = df.groupby('hierarchical_cluster').mean()
print("Hierarchical Cluster Analysis:\n", hierarchical_cluster_analysis)

# Analyze clusters for DBSCAN
df['dbscan_cluster'] = dbscan_labels
dbscan_cluster_analysis = df[df['dbscan_cluster'] != -1].groupby('dbscan_cluster').mean()
print("DBSCAN Cluster Analysis:\n", dbscan_cluster_analysis)
```
