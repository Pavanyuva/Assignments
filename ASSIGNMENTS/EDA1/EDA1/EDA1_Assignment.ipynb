#3.	Data Visualization:

histograms distributions of various numerical variables

```python
df_num=df.select_dtypes(include=['number'])
```

```python
df_num.hist(figsize=(12,12),grid=False)
plt.show()
```

**Boxplots to detect outliers**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplots for All Variables")
plt.xticks(rotation=90)
plt.show()
```

**bar charts display the frequency of categories for categorical variables**

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='NSP', data=df)
plt.title('Bar Chart of Health Status')
plt.xlabel('Health Status')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

Categories: ['Normal', 'Warning', 'Critical']

```python
df['HealthStatus'] = pd.cut(df['LB'], bins=[100, 120, 140, 160], labels=['Normal', 'Warning', 'Critical'])

plt.figure(figsize=(8, 6))
df['HealthStatus'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Bar Chart of Health Status')
plt.xlabel('Health Status')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

Generate scatter plots heatmaps to explore relationships between pairs of variables

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['LB'], df['ASTV'], alpha=0.5)
plt.title('Scatter Plot between LB and ASTV')
plt.xlabel('LB (Baseline Fetal Heart Rate)')
plt.ylabel('ASTV (Abnormal Short Term Variability)')
plt.show()
```

Corrlection Heatmap

```python
df = pd.get_dummies(df, columns=['HealthStatus'])

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Cardiographic Data")
plt.show()
```

Employ advanced visualization techniques like pair plots, or violin plots for deeper insights

Pairplot to explore relationships between variables

```python
sns.pairplot(df)
plt.title("Pairplot of Cardiographic Data")
plt.show()
```

Violin Plots

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df)
plt.title("Violin Plots for All Variables")
plt.xticks(rotation=90)
plt.show()
```
