**Task 2: Dimensionality Reduction with PCA:**
1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.
2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
3.	Transform the original dataset into the principal components.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Standardizing the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
print("First few rows of the standardized dataset:")
print(df_scaled[:5])
```

```python
# Applying PCA
pca = PCA()
pca.fit(df_scaled)

# Scree Plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(df.columns) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
# Cumulative explained variance
cum_var_exp = pca.explained_variance_ratio_.cumsum()

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(df.columns) + 1), cum_var_exp, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# Finding the optimal number of components
for i, var in enumerate(cum_var_exp):
    if var >= 0.90:
        print(f"Optimal number of components: {i+1} (Explained variance: {var:.2f})")
        break
```

```python
# Set the number of components
n_components = i+1
pca_final = PCA(n_components=n_components)

# Transform the data into the principal components
df_pca = pca_final.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=[f'PC{i+1}' for i in range(n_components)])

print("First few rows of the dataset transformed into Principal Components:")
print(df_pca.head())
```

**Task 3: Clustering with Original Data:**
1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.
2.	Visualize the clustering results using appropriate plots.
3.	Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index.


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Define the number of clusters (K)
k = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(df_scaled)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Add cluster labels to the original dataframe
df['Cluster'] = cluster_labels

print("Cluster labels assigned to each data point:")
print(df.head())
```

```python
# Reduce data to 2D using PCA for visualization
pca_2d = PCA(n_components=2)
df_pca_2d = pca_2d.fit_transform(df_scaled)

# Convert PCA results into a DataFrame
df_pca_2d = pd.DataFrame(df_pca_2d, columns=['PC1', 'PC2'])
df_pca_2d['Cluster'] = cluster_labels

# Visualizing clusters in a 2D PCA plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=df_pca_2d, palette='Set2', s=100)
plt.title('K-means Clustering Results in 2D PCA')
plt.show()
```

```python
# Calculate silhouette score
silhouette_avg = silhouette_score(df_scaled, cluster_labels)
print(f'Silhouette Score: {silhouette_avg:.2f}')

# Calculate Davies–Bouldin Index
davies_bouldin_avg = davies_bouldin_score(df_scaled, cluster_labels)
print(f'Davies–Bouldin Index: {davies_bouldin_avg:.2f}')
```

**Task 4: Clustering with PCA Data:**
1.	Apply the same clustering algorithm to the PCA-transformed dataset.
2.	Visualize the clustering results obtained from PCA-transformed data.
3.	Compare the clustering results from PCA-transformed data with those from the original dataset.


```python
# K-means clustering on PCA-transformed data
kmeans_pca = KMeans(n_clusters=k, random_state=42)
kmeans_pca.fit(df_pca)

# Get the cluster labels
cluster_labels_pca = kmeans_pca.labels_

# Add the PCA-based cluster labels
df_pca['Cluster'] = cluster_labels_pca

print("Cluster labels for PCA-transformed data:")
print(df_pca.head())
```

```python
# Visualize the clustering results in 2D PCA plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=df_pca, palette='Set1', s=100)
plt.title('K-means Clustering Results on PCA-Transformed Data in 2D')
plt.show()
```

```python
# Calculate silhouette score for PCA-transformed data
silhouette_avg_pca = silhouette_score(df_pca.iloc[:, :-1], cluster_labels_pca)
print(f'Silhouette Score (PCA-transformed data): {silhouette_avg_pca:.2f}')

# Calculate Davies–Bouldin Index for PCA-transformed data
davies_bouldin_avg_pca = davies_bouldin_score(df_pca.iloc[:, :-1], cluster_labels_pca)
print(f'Davies–Bouldin Index (PCA-transformed data): {davies_bouldin_avg_pca:.2f}')
```
